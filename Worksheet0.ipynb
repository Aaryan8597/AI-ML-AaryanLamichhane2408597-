## Task-1 Creating a Python program that converts between different units of measurement

```python
#function for unit conversion
def convert_units(conversion_type, value):
  """
    Converts units for length, weight, and volume.
    conversion_type (int): type of conversion selected by the user
    value (float): value entered by the user
    returns
    float: converted result
    """
  if conversion_type == 1: #meters to feet
    return value * 3.28084
  elif conversion_type == 2: #feet to meters
    return value / 3.28084
  elif conversion_type == 3: #kilograms to pounds
    return value * 2.20462
  elif conversion_type == 4: #pounds to kilograms
      return value / 2.20462
  elif conversion_type == 5: #liters to gallons
      return value * 0.264172
  elif conversion_type == 6: #gallons to liters
      return value / 0.264172
  else:
      return None

print("Unit Conversion Program")
print("\nLength Conversion")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("\nWeight Conversion")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("\nVolume Conversion")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    conversion_type = int(input("\nEnter conversion choice (1-6): "))
    value = float(input("Enter the value to convert: "))

    result = convert_units(conversion_type, value)
    if conversion_type >= 1 and conversion_type <= 6:
      print("Converted value:", result)
    else:
      print("Invalid conversion type.")

except ValueError:
    print("Error: Please enter a numeric value.")
```

## Task-2 Creating a Python program that performs various mathematical operations on a list of numbers.

```python
def find_sum(numbers):
    """
    Calculate the sum of a list of numbers.
    numbers (list): list of numeric values
    Returns:
    float: sum of all numbers
    """
    return sum(numbers)

def find_average(numbers):
    """
    Calculate the average of a list of numbers.
    numbers (list): list of numeric values
    Returns:
    float: average value
    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Find the maximum number in the list.
    numbers (list): list of numeric values
    Returns:
    float: largest number in the list
    """
    return max(numbers)

def find_min(numbers):
    """
    Find the minimum number in the list.
    numbers (list): list of numeric values

    Returns:
    float: smallest number in the list
    """
    return min(numbers)

print("Mathematical Operations Program")
print("1. Find Sum")
print("2. Find Average")
print("3. Find Maximum")
print("4. Find Minimum")
try:
    operation = int(input("Choose an operation (1-4): "))
    numbers_input = input("Enter numbers separated by spaces: ")
    numbers_list = [float(num) for num in numbers_input.split()]
    if len(numbers_list) == 0:
        print("Error: The list is empty.")
    elif operation == 1:
        result = find_sum(numbers_list)
        print("Sum:", result)
    elif operation == 2:
        result = find_average(numbers_list)
        print("Average:", result)
    elif operation == 3:
        result = find_max(numbers_list)
        print("Maximum:", result)
    elif operation == 4:
        result = find_min(numbers_list)
        print("Minimum:", result)
    else:
        print("Invalid operation choice.")
except ValueError:
    print("Error: Please enter valid numeric values.")
```

# Exercise on Nested List

## Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.

```python
def flatten(first):
  flat_list = []
  for sublist in first:
    flat_list.extend(sublist)
  return flat_list

nested_list = [[1, 2], [3, 4], [5, 6]]
result = flatten(nested_list)
print("Original nested list:", nested_list)
print("Flattened list:", result)
```

## Write a Python function that extracts a specific element from a nested list given its indices.

```python
def access_nested_element(first, indices):
  element = first
  for i in indices:
    element = element[i]
  return element

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)
print("Nested list:", nested_list)
print(f"Element at indices {indices}:", result)
```

## Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

```python
def sum_nested(first):
  total = 0
  for item in first:
    try:
      total += sum_nested(item)
    except TypeError:
      total += item
  return total

nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)
print("Nested list:", nested_list)
print("Sum of all elements:", result)
```

## Write a Python function that removes all occurrences of a specific element from a nested list.

```python
def remove_element(first, elem):
  new_list = []
  for item in first:
      try:
          new_list.append(remove_element(item, elem))
      except TypeError:
          if item != elem:
              new_list.append(item)
  return new_list

first = [[1, 2], [3, 2], [4, 5]]
result = remove_element(first, 2)

print("Original list:", first)
print("Modified list:", result)
```

## Write a Python function that finds the maximum element in a nested list (regardless of depth).

```python
def find_max(lst):
    max_value = 0
    for item in lst:
        if type(item) == list:
            value = find_max(item)
        else:
            value = item
        if value > max_value:
            max_value = value
    return max_value

numbers = [[1, 2], [3, [4, 5]], 6]
result = find_max(numbers)
print("Maximum element:", result)
```

## Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if type(item) == list:
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

lst = [[1, 2], [2, 3], [2, 4]]
result = count_occurrences(lst, 2)
print("Occurrences:", result)
```

## Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if type(item) == list:
            result += deep_flatten(item)
        else:
            result.append(item)
    return result

lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(lst))
```

## Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):
    total_sum = 0
    total_count = 0
    for item in lst:
        if type(item) == list:
            sub_sum, sub_count = average_nested(item)
            total_sum += sub_sum
            total_count += sub_count
        else:
            total_sum += item
            total_count += 1
    return total_sum, total_count

numbers = [[1, 2], [3, 4], [5, 6]]
total_sum, total_count = average_nested(numbers)
average = total_sum / total_count
print("Average:", average)
```

## Problem-4 Matrix Operations

```python
# Prove following with Numpy:
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
print("Matrix A:\n", A)
print("Matrix B:\n", B)

# proving A.A−1 = I.
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("\n1. A * A^-1:\n", identity)
print("Is approximately identity matrix:", np.allclose(identity, np.eye(2)))

# proving AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB:\n", AB)
print("BA:\n", BA)
print("AB equals BA?", np.array_equal(AB, BA))

#proving (AB) T = BT AT
AB_transpose = AB.T
B_transpose_A_transpose = np.dot(B.T, A.T)
print("\n(AB)^T:\n", AB_transpose)
print("B^T * A^T:\n", B_transpose_A_transpose)
print("Are they equal?", np.allclose(AB_transpose, B_transpose_A_transpose))
```

```python
# solving the following system of linear equation using inverse methods
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])
const_matrix = np.array([-1, -3, 9])

coeff_inv = np.linalg.inv(coeff_matrix)
solution = np.dot(coeff_inv, const_matrix)
print("\nSolution using inverse method:", solution)
```
